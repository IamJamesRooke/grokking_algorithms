```python
def countdown(i):
    print(i)
    if i <= 0: # base case, when function not calling self again
        return
    else:
        countdown(i-1) # recursive case, when the function calls itself

countdown(5)
```

# Factorial

```python
def fact(x):
    if x == 1:
        return 1
    else:
        return fact(x-1) * x

fact(3)
```
